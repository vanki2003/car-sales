```python
numerical_columns = df.select_dtypes(include = ['int64', 'float64']).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f'{feature} | Skewness: {round(df[feature].skew(), 1)}')
    
plt.tight_layout()
plt.show()
```

## Price Analysis

Analyze the distribution of selling prices. How do factors like year, make, model, condition, and mileage affect the selling price?

```python
fig, axs = plt.subplots(3, 1, figsize=(10,5))

sns.scatterplot(ax=axs[0], data=df, x='sellingprice', y='year')

sns.scatterplot(ax=axs[1], data=df, x='sellingprice', y='condition')

sns.scatterplot(ax=axs[2], data=df, x='sellingprice', y='odometer')

plt.tight_layout()
plt.show()
```
