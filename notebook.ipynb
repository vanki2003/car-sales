```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
sns.set_style("darkgrid")
sns.set_theme("paper")
sns.color_palette("mako", as_cmap=True)
```

```python
pwd = os.getcwd()
```

```python
# loading the data
df = pd.read_csv(pwd + '/car_prices.csv')
df
```

```python
df.describe(include=object)
```

```python
numerical_columns = df.select_dtypes(include = ['int64', 'float64']).columns

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f'{feature} | Skewness: {round(df[feature].skew(), 1)}')
    
plt.tight_layout()
plt.show()
```

```python
sns.regplot(data=df, x='sellingprice', y='mmr', line_kws={"color":"r","lw":2})
```

```python
df['odometer'].describe()
```

```python
high_mileage = df[(df['odometer'] > 51085) & (df['odometer'] < 999999)]
high_mileage
```

```python
low_mileage = df[(df['odometer'] < 51085)]
low_mileage
```

```python
high_mileage.groupby('make')['sellingprice'].mean().plot(kind='barh', figsize=(15, 10))
plt.xlabel('Average Selling Price (USD)')
plt.ylabel('Make')
plt.title('Average Selling Price by Make (High Mileage)')
plt.show()

low_mileage.groupby('make')['sellingprice'].mean().plot(kind='barh', figsize=(15, 10))
plt.xlabel('Average Selling Price (USD)')
plt.ylabel('Make')
plt.title('Average Selling Price by Make (Low Mileage)')
plt.show()
```
